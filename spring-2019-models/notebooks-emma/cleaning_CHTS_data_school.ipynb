## Step 3. Convert list of locations to list of trips

### 3a. Make trip IDs and list home-work and work-home modes by trip ID

```python
def count_place(df):
    return df.join(df['place'].str.get_dummies()
                  .cumsum().add_suffix('_Count')).head(n=15)

trips = place5.groupby('HHPER', group_keys=False).apply(count_place)

trips.head()
```

```python
neverhome = trips[trips.home_Count.isnull()]

neverathome = neverhome.HHPER.unique()

neverhome.HHPER.nunique()
```

```python
trips = trips[~trips.HHPER.isin(neverathome)]
```

```python
# personsnew3["home_Count"].fillna(0, inplace=True)

# personsnew3["no_Count"].fillna(0, inplace=True)

trips["trip_ID"] = trips["school_Count"].map(str) + trips["home_Count"].map(str)

trips["HHPERTRIP"] = trips["HHPER"].map(str) + trips["trip_ID"].map(str)

trips['count'] = trips.groupby('HHPERTRIP').cumcount() + 1

trips.head(n=20)
```

```python
trips['count'].max()
```

```python
trips1 = trips.pivot(index='HHPERTRIP', columns='count', values='MODE_NEW').reset_index()

trips1.columns = ['HHPERTRIP', 'MODE1', 'MODE2', 'MODE3', 'MODE4', 'MODE5', 'MODE6', 'MODE7', 
                       'MODE8', 'MODE9', 'MODE10', 'MODE11', 'MODE12', 'MODE13', 'MODE14', 'MODE15']

trips1.head()
```

```python
trips = pd.merge(trips, trips1, on='HHPERTRIP')

trips.head()
```

### 3b. Isolate rows that identify start of home-school or school-home trips.

```python
#List of all departures from home or school by students.
homeschooltrips = trips.loc[trips['place'].isin(["home","school"])]
```

```python
homeschooltrips['trip_ET'] = homeschooltrips.groupby('HHPER', group_keys=False).ST_num.shift(-1)
```

```python
homeschooltrips = homeschooltrips[['HHPER', 'HHPERTRIP','place', 'ST_num','ET_num','trip_ET','MODE_NEW',
                               'MODE1', 'MODE2', 'MODE3', 'MODE4', 'MODE5', 'MODE6', 'MODE7', 
                           'MODE8', 'MODE9', 'MODE10', 'MODE11', 'MODE12', 'MODE13', 'MODE14','MODE15']].copy()

homeschooltrips.head(n=10)
```

### 3c. Identify trips where the departure place is not the same as the one before it. This captures 'from home' trips that end at school and 'from school' trips that end at home. 

### However, it also captures end-of-day 'from home' trips that don't end at school and 'from school' trips that don't end at home.

```python
#Taking last trips where the place is not the same as the one before it and/or the person is not the same as the one before it.

tripsI = homeschooltrips.loc[(homeschooltrips.HHPER != homeschooltrips.HHPER.shift(-1)) | (homeschooltrips.place != homeschooltrips.place.shift(-1))]

tripsI.head(n=10)
```

### 3d. Identify end-of-day 'from home' trips that end at school, as well as end-of-day 'from school' trips that don't end at home.

```python
#Identifying people whose last home or school location is school.

tripsJ = homeschooltrips.loc[homeschooltrips.HHPER != homeschooltrips.HHPER.shift(-1)]

tripsJ.groupby(['place']).count()
```

```python
tripsK = tripsJ.loc[tripsJ['place'].isin(["school"])]
```

#### Ends day at school (late-night studying). End-of-day 'from home' trip ends at school:

```python
tripsL = tripsK[(tripsK['ET_num'] >= 2.98) & (tripsK['ET_num'] <= 2.985)]

nightshift = tripsL.HHPER.unique()

tripsL.HHPER.nunique()
```

#### Leaves school but does not go home at end of day (to airport, second home, or another location). End-of-day 'from school' trip does not end at home:

```python
tripsM = tripsK[(tripsK['ET_num'] <= 2.98) | (tripsK['ET_num'] >= 2.985)]

dontgohome = tripsM.HHPERTRIP.unique()

tripsM.HHPER.nunique()
```

### 3e. Get rid of last 'from home' trip for all those except late-night students, and get rid of last school activity for late-night students.

```python
#Selecting all but people whose last 'from home' trip is to school. 
tripsII = tripsI[~tripsI.HHPER.isin(nightshift)] 

#Selecting people whose last 'from home' trip is to school.
tripsIII = tripsI[tripsI.HHPER.isin(nightshift)]

#Get rid of last work activity for late-night students.
tripsIII = tripsIII[(tripsIII['ET_num'] <= 2.9833) | (tripsIII['ET_num'] >= 2.9834)]

tripsIII.head()

print (tripsII.HHPER.nunique())

print (tripsIII.HHPER.nunique())
```

```python
#Getting rid of last 'from home' trip for all those except people going to school.

tripsN = tripsII.groupby('HHPER').apply(lambda x: x.iloc[:-1] if (x.iloc[-1, x.columns.get_loc('place')] == "home") else x).reset_index(drop=True)

tripsN.HHPER.nunique()
```

```python
#Adding back people going to work as last 'from home' trip.
tripsP = pd.concat([tripsN,tripsIII])
```
