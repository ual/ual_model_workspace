```python
zones_2010 = pd.read_csv('./data/zones_w_access_vars-2010.csv')
zones_2025 = pd.read_csv('./data/zones_w_access_vars-2025-b-lt.csv')
zones_2040 = pd.read_csv('./data/zones_w_access_vars-2040-b-lt.csv')
```

```python
gdf = gpd.read_file('/home/data/spring_2019/base/Transportation_Analysis_Zones.shp')
```

```python
gdf_2010 = gdf.merge(
    zones_2010[['total_jobs_gen_tt_CAR_30', 'total_jobs_gen_tt_WALK_TRANSIT_30']],
    left_on='taz1454', right_index=True)
gdf_2025 = gdf.merge(
    zones_2025[['total_jobs_gen_tt_CAR_30', 'total_jobs_gen_tt_WALK_TRANSIT_30']],
    left_on='taz1454', right_index=True)
gdf_2040 = gdf.merge(
    zones_2040[['total_jobs_gen_tt_CAR_30', 'total_jobs_gen_tt_WALK_TRANSIT_30']],
    left_on='taz1454', right_index=True)
```

```python
def car_vs_transit_plots(i):
    tts = [15, 30, 60]
    tt = tts[i]
    gdf_merged = gdf.merge(
    zones_2010[['total_jobs_gen_tt_CAR_30', 'total_jobs_gen_tt_CAR_15', 'total_jobs_gen_tt_CAR_60',
           'total_jobs_gen_tt_WALK_TRANSIT_30', 'total_jobs_gen_tt_WALK_TRANSIT_15', 'total_jobs_gen_tt_WALK_TRANSIT_60'
    ]], left_on='taz1454', right_index=True)
    
    kwargs = {
#         'scheme': 'user_defined', 'classification_kwds': {'bins': [1e5, 2e5, 3e5, 4e5, 6e5]},
        'scheme': 'quantiles',
        'legend_kwds': {'loc': 3}
    }
    fig.suptitle('Jobs within {0} minutes'.format(tt), fontsize=15, y=.85)
    gdf_merged.plot(
        column='total_jobs_gen_tt_CAR_{0}'.format(tt),
        ax=axarr[0], legend=True, **kwargs)
    _ = axarr[0].set_title('Car', fontsize=14)
    axarr[0].axes.get_yaxis().set_ticks([])
    axarr[0].axes.get_xaxis().set_ticks([])
    
    gdf_merged.plot(
        column='total_jobs_gen_tt_WALK_TRANSIT_{0}'.format(tt),
        ax=axarr[1], legend=True, **kwargs)
    _ = axarr[1].set_title('Walk to Transit', fontsize=14)
    axarr[1].axes.get_yaxis().set_ticks([])
    axarr[1].axes.get_xaxis().set_ticks([])
    
def sim_year_plots(i):
    gdfs = [gdf_2010, gdf_2025, gdf_2040]
    years = [2010, 2025, 2040]
    year = years[i]
    gdf = gdfs[i]
    tt = 30
    kwargs = {
#         'scheme': 'user_defined', 'classification_kwds': {'bins': [2e5, 4e5, 6e5, 10e5, 15e5]},
        'scheme': 'quantiles',
        'legend_kwds': {'loc': 3}
    }
    fig.suptitle('Jobs within 30 minutes -- {0}'.format(year), fontsize=15, y=.85)
    gdf.plot(
        column='total_jobs_gen_tt_CAR_{0}'.format(tt),
        ax=axarr[0], legend=True, **kwargs)
    _ = axarr[0].set_title('Car', fontsize=14)
    axarr[0].axes.get_yaxis().set_ticks([])
    axarr[0].axes.get_xaxis().set_ticks([])
    
    gdf.plot(
        column='total_jobs_gen_tt_WALK_TRANSIT_{0}'.format(tt),
        ax=axarr[1], legend=True, **kwargs)
    _ = axarr[1].set_title('Walk to Transit', fontsize=14)
    axarr[1].axes.get_yaxis().set_ticks([])
    axarr[1].axes.get_xaxis().set_ticks([])
    
def init_plot():
    fig.suptitle(' ', fontsize=15, y=.85)
    _ = gdf.plot(axarr[0].plot())
    _ = axarr[1].plot()
```

```python
%matplotlib notebook
fig, axarr = plt.subplots(1, 2, figsize=(16,9))
maps_ani = animation.FuncAnimation(fig, car_vs_transit_plots, 3)
```

```python
%matplotlib notebook
fig, axarr = plt.subplots(1, 2, figsize=(16,9))
maps_ani = animation.FuncAnimation(fig, sim_year_plots, 3)
```
