```python
import pandas as pd
import seaborn as sns
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv('../data/sampling_results_v2.csv', index_col=0)
df.index.name = 'n_alts'
df.reset_index(inplace=True)
```

```python
df = df.sort_values('n_alts')
```

```python
fig, ax = plt.subplots(figsize=(12,8))
ax.plot(df['n_alts'], df['avg_max'], label='avg. est. $p_{max}(i)$', marker='o')
ax.legend(fontsize=16)
ax.set_xlabel('choice set size', fontsize=15)
```

```python
fig, ax = plt.subplots(figsize=(12,8))
ax.plot(df['n_alts'], df['avg_max'] / df['avg_median']/df['avg_stddev']), label='asdf', marker='o')
ax.legend(fontsize=16)
ax.set_xlabel('choice set size', fontsize=15)
```
